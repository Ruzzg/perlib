```python
data_l  = data_l[-1000:]
data_l = data_l.fillna(0)
```

```python
data_l.tail()
```

```python
from perlib.core.models.smodels import models as armodels
aR_info.modelname = "arima"
aR_info.forecastNumber = 7
aR_info.period = "hourly"
aR_info.forcastingStartDate = "2018-12-31 00:00:00"
ar = armodels(aR_info)
```

```python
from perlib.core.train import  sTrain
train = sTrain(dataFrame=data_l,object=ar)
res = train.fit()
```

```python
from perlib.core.tester import sTester
r = sTester(dataFrame=data_l,object=ar,path="Data-arima-2023-01-09-18-56-54.pkl",metric=["mape","mae","mse"])
```
